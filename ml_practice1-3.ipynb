```python
#練習問題3
import numpy as np
data = np.array(((20,30), (60,75), (80,30), (90,70), (95,90), (40,60), (80,90), (30,40), (25,55), (35,25), (80,45), (50,10), (25,80)))
y = np.array((0,1,0,1,1,1,1,0,0,0,1,0,1))
w = (0.1, 0.3, -20)

def classify(w,x):
    if w[0]*x[0] + w[1]*x[1] + w[2] >= 0:
        return 1
    else:
        return 0

wrong = 0
for x, correct in zip (data, y):
    print("x = ", x, ", correct =  ", correct)
    pred = classify(w,x)
    wrong += np.abs(pred - correct)
print("wrong = ", wrong)
```
