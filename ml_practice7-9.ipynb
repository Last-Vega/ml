```python
#練習問題9
import numpy as np
W1 = np.array([[5,3], [2,-1], [4,2]])
x1 = np.array([3,2])
def softmax(a):
    u = np.sum(np.exp(a))
    return np.exp(a)/u

def classify(W,x):
    return softmax(np.matmul(W,x))

print(classify(W1,x1))
print(np.sum(classify(W1,x1)))
```
