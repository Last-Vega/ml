```python
#練習問題6
import numpy as np
data = np.array(((20,30), (60,75), (80,30), (90,70), (95,90), (40,60), (80,90), (30,40), (25,55), (35,25), (80,45), (50,10), (25,80)))
w = (0.1, 0.3, -20)
new_data = []
for i in range(len(data)):
    elem1 = data[i][0]
    elem2 = data[i][1]
    new_data.append([elem1, elem2, 1])

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

for i in range(len(new_data)):
    activation = np.matmul(w, new_data[i])
    probability = sigmoid(activation) * 100
    print(data[i],"のとき:",  round(probability, 2), "%")
```
